```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
df = pd.read_csv(url, header=None)

# Add column names
column_names = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'income']
df.columns = column_names
```

```python
# Display first 5 rows of data
df.head(5)
```

```python

# Check data types and missing values
df.info()
```

```python

# Check summary statistics
df.describe()
```

```python

# Check unique values in categorical columns
print(df['workclass'].value_counts())
print(df['occupation'].value_counts())
print(df['native-country'].value_counts())
```

Now handling missing values

```python
# Identify missing values
df.isnull().sum()
```

```python
# Replace '?' with NaN
df.replace('?', np.nan, inplace=True)
```

```python

# Check missing values again
df.isnull().sum()
```

```python

# Check data info after handling missing values
df.info()
```

```python
# Plot histogram of numerical columns
df.hist(figsize=(10, 10), grid=False)
plt.show()
```

```python
# Plot bar charts of categorical columns
sns.set(style="whitegrid")
# Create a separate figure for each plot
fig1, ax1 = plt.subplots(figsize=(8, 6))
fig2, ax2 = plt.subplots(figsize=(8, 6))
fig3, ax3 = plt.subplots(figsize=(8, 6))
fig4, ax4 = plt.subplots(figsize=(8, 6))
fig5, ax5 = plt.subplots(figsize=(8, 6))
fig6, ax6 = plt.subplots(figsize=(8, 6))

# Plot the bar charts
sns.countplot(x='workclass', data=df, ax=ax1)
sns.countplot(x='education', data=df, ax=ax2)
sns.countplot(x='marital-status', data=df, ax=ax3)
sns.countplot(x='income', data=df, ax=ax4)
sns.countplot(x='relationship', data=df, ax=ax5)
sns.countplot(x='race', data=df, ax=ax6)

# Set titles for each plot
ax1.set_title('Workclass')
ax2.set_title('Education')
ax3.set_title('Marital Status')
ax4.set_title('Income')
ax5.set_title('Relationship')
ax6.set_title('Race')

# Rotate x-axis labels by 90 degrees
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=90)
ax2.set_xticklabels(ax2.get_xticklabels(), rotation=90)
ax3.set_xticklabels(ax3.get_xticklabels(), rotation=90)
ax4.set_xticklabels(ax4.get_xticklabels(), rotation=90)
ax5.set_xticklabels(ax5.get_xticklabels(), rotation=90)
ax6.set_xticklabels(ax6.get_xticklabels(), rotation=90)

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plots
plt.show()
```

```python
# Plot boxplots of numerical columns
sns.set(style="whitegrid")
fig, axes = plt.subplots(2, 2, figsize=(15, 15))
fig.tight_layout(pad=5.0)
sns.boxplot(x='income', y='age', data=df, ax=axes[0, 0])
sns.boxplot(x='income', y='education-num', data=df, ax=axes[0, 1])
sns.boxplot(x='income', y='capital-gain', data=df, ax=axes[1, 0])
sns.boxplot(x='income', y='capital-loss', data=df, ax=axes[1, 1])
plt.show()
```

```python
# Plot correlation matrix
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()
```
