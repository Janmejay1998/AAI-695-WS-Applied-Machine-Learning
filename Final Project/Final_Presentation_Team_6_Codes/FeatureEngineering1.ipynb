```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
csfont = {'fontname':'Times New Roman'}
```

```python
# Load the CSV dataset
df = pd.read_csv('adult.csv')
df.head(3)
```

```python
# Check for Null Data
df.isnull().sum()
```

```python
df['income']=df['income'].map({'<=50K': 0, '>50K': 1})
```

```python
# Testing if any insight in any selected feature
feature = 'native-country'#'education'
print(sorted(list(df[feature].unique())))
print(len(list(df[feature].unique())))
ax = sns.catplot(x = feature,y= 'income' ,data = df,kind="bar")
ax.set_xticklabels(rotation=90)#, **csfont)
```

```python
# Testing if any insight in any selected feature
feature = 'marital-status'#'education'
print(sorted(list(df[feature].unique())))
print(len(list(df[feature].unique())))
ax = sns.catplot(x = feature,y= 'income' ,data = df,kind="bar")
ax.set_xticklabels(rotation=90, **csfont)
```

```python
# Create Married Column - Binary Yes(1) or No(0)
df["marital-status"] = df["marital-status"].replace(['Never-married','Divorced','Separated','Widowed','Married-spouse-absent'], 0)
df["marital-status"] = df["marital-status"].replace(['Married-civ-spouse','Married-AF-spouse'], 1)
```

```python
# Testing if any insight in any selected feature
feature = 'relationship'#'education'
print(sorted(list(df[feature].unique())))
print(len(list(df[feature].unique())))
ax = sns.catplot(x = feature,y= 'income' ,data = df,kind="bar")
ax.set_xticklabels(**csfont)
```

```python
# Testing if any insight in any selected feature
feature = 'relationship'#'education'
print(sorted(list(df[feature].unique())))
print(len(list(df[feature].unique())))
ax = sns.catplot(x = feature,y= 'income' ,data = df,kind="bar")
ax.set_xticklabels(rotation=90, **csfont)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axs = plt.subplots(3, 1, figsize=(10, 20))
features = ['native-country', 'marital-status', 'relationship']

for ax, feature in zip(axs.flat, features):
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
#     ax.set_title(feature)
    sns.barplot(x=feature, y='income', data=df, ax=ax)

plt.tight_layout()
plt.show()
```
